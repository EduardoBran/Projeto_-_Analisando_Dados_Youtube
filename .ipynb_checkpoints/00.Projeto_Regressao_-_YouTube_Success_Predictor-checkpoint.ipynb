### Contexto

A empresa **Global Media Insights** é uma consultoria especializada em análises de mídia e entretenimento.

Em 2023, a empresa foi contratada pela **Creative Content Network (CCN)**, uma agência de talentos que gerencia alguns dos maiores criadores de conteúdo do YouTube, para analisar e prever o sucesso de seus canais. A agência deseja otimizar suas estratégias de gestão e marketing, investindo em criadores de conteúdo com maior potencial de crescimento.

A **Creative Content Network (CCN)** gerencia um portfólio diversificado de canais no YouTube, abrangendo diversas categorias e regiões do mundo. Com a crescente competição e a evolução constante das preferências do público, a agência percebeu a necessidade de uma análise mais aprofundada baseada em dados para tomar decisões estratégicas.

<br>

## Objetivo Principal do Projeto

- O objetivo principal deste projeto é **desenvolver um modelo preditivo que permita à Creative Content Network (CCN) prever o número de novos inscritos em canais do YouTube para o próximo mês**. Com isso, a agência poderá tomar decisões mais informadas sobre quais criadores apoiar mais intensamente e como direcionar seus esforços de marketing.

#### O que a empresa espera receber

- A Creative Content Network (CCN) espera receber um conjunto de ferramentas e análises que possibilitem prever o crescimento dos inscritos em canais do YouTube e fornecer insights estratégicos para otimizar seus esforços de marketing e gestão.

<br>

### Entregáveis do Projeto

#### Modelo Preditivo
> Um modelo de machine learning treinado para prever o número de novos inscritos em um canal do YouTube nos próximos 30 dias com base nas características do canal.

#### Interface Interativa
> Uma aplicação web interativa onde os usuários poderão inserir características específicas de um canal do YouTube e obter a previsão do número de novos inscritos para os próximos 30 dias.
A interface permitirá a entrada de variáveis como categoria do canal, número de visualizações, frequência de uploads, país de origem, e outras características relevantes.

#### Relatório de Análise e Insights
> Insights acionáveis sobre as características que mais influenciam o crescimento de inscritos.
Recomendações estratégicas baseadas nos resultados do modelo preditivo.

#### Dashboard de Visualização
> Um dashboard interativo que permite aos usuários explorar os dados e visualizar tendências, padrões e previsões.
Gráficos e visualizações que mostram a distribuição de inscritos, visualizações de vídeos, categorias mais populares, e outras métricas relevantes.

### Perguntas de Negócios Adicionais

- 1. Quais características dos canais (como categoria, frequência de uploads, país de origem, etc.) estão mais fortemente correlacionadas com o aumento de inscritos?
- 2. Existe uma correlação significativa entre o número de visualizações de vídeos e o crescimento de inscritos no YouTube?
- 3. Como diferentes categorias de conteúdo impactam o crescimento dos canais no YouTube?
- 4. Quais estratégias de conteúdo (como tipos de vídeos, duração dos vídeos, etc.) são mais eficazes para maximizar o crescimento de inscritos?
- 5. Como o desempenho dos canais nos últimos 30 dias pode ser utilizado para prever o crescimento futuro?

### Conclusão

Com essa abordagem, a Creative Content Network (CCN) terá uma ferramenta poderosa para prever o crescimento de inscritos em canais do YouTube e tomar decisões estratégicas baseadas em dados. Isso permitirá à agência otimizar seus lançamentos e estratégias de marketing, aumentando as chances de sucesso de seus criadores de conteúdo no mercado competitivo de hoje.
