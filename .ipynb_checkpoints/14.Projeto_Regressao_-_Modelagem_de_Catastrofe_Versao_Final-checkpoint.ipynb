### Motivo da Exclusão de Variáveis

Aqui está uma breve explicação para a exclusão de cada variável:

- *Dis No*: Número de identificação do desastre, não relevante para a análise preditiva.
- *Glide*: Identificação global do desastre, não necessária para a modelagem.
- *Disaster_Group*: Essa variável possui apenas um valor constante e assim não fornece informação diferenciada para o modelo preditivo, pois não contribui para a variabilidade dos dados.
- *Disaster_Subsubtype*: Muitos valores ausentes (apenas 496 de 4863 não são nulos), tornando-a difícil de usar de forma eficaz.
- *Event Name*: Nome do evento, considerado não relevante para a análise preditiva.
- *Country*: A variável Country oferece uma categorização mais detalhada, mas é amplamente representada pela variável Continent, que agrupa países em categorias geográficas mais amplas. Outro problema é: com 200 categorias únicas, Country pode introduzir sparsity nos dados, dificultando a modelagem e aumentando o risco de overfitting
- *ISO*: Código ISO do país, não necessário, pois a variável Continent já engloba os dados de localização.
- *Location*: A variável Location apresenta 4163 categorias únicas, o que pode introduzir muita granularidade e complexidade ao modelo, dificultando a interpretação e a generalização. Além disso, informações geográficas detalhadas já são representadas de maneira mais consolidada e útil pelas variáveis Country e Continent
- *Origin*: Muitos valores ausentes (1289 de 4863).
- *Associated_Dis* e *Associated_Dis2*: Muitos valores ausentes (1538 e 435 de 4863, respectivamente).
- *OFDA_Response*: Muitos valores ausentes (799 de 4863).
- *Appeal*: Muitos valores ausentes (1069 de 4863).
- *Declaration*: Muitos valores ausentes (1401 de 4863).
- *Latitude*: Latitude da localização do desastre, muitos valores ausentes e localização geográfica já representada por outras variáveis.
- *Longitude*: Longitude da localização do desastre, muitos valores ausentes e localização geográfica já representada por outras variáveis.
- *Local Time*: Tempo local do evento, não relevante para a análise preditiva.
- *River Basin*: Bacia hidrográfica afetada, muitos valores ausentes e considerada menos relevante.
- *Aid Contribution*: Contribuição de ajuda em milhares de dólares, muitos valores ausentes e não é diretamente relevante.
- *Dis_Mag_Value*: Muitos valores ausentes.
- *Admin1 Code*: Código administrativo 1, muitos valores ausentes e granularidade das unidades administrativas menos relevante.
- *Admin2 Code*: Código administrativo 2, muitos valores ausentes e granularidade das unidades administrativas menos relevante.
- *Adm Level*: Muitos valores ausentes (2771 de 4863), e pode ser menos relevante.
- *Geo Locations*: Localizações geográficas afetadas, muitos valores ausentes e localização já representada por outras variáveis.
- *Reconstruction Costs*: Custo de reconstrução em milhares de dólares, muitos valores ausentes.
- *No_Injured e No_Affected*: Ambas as variáveis têm uma alta proporção de valores ausentes. A presença de muitos valores extremos (outliers) torna a imputação mais desafiadora e pode não resultar em estimativas confiáveis.
- *No_Homeless*: Muitos valores ausentes, dificultando a utilização eficaz desta variável na modelagem.
- *Insured Damages*: Danos segurados em milhares de dólares, muitos valores ausentes e não diretamente relevante para o modelo preditivo focado em danos totais.
- *CPI*: Índice de Preços ao Consumidor, pode não ser diretamente relevante para a predição de danos causados por catástrofes naturais.

```python
# Função para plotar histogramas e boxplots em um único plot
def plot_variaveis_ausentes(data, variaveis):
    num_variaveis = len(variaveis)
    num_cols = 2  # Histograma e Boxplot lado a lado
    num_rows = (num_variaveis + 1) // num_cols  # Dividindo por 2 porque cada variável terá dois gráficos
    
    fig, axs = plt.subplots(num_rows, num_cols*2, figsize=(20, num_rows * 6))
    fig.suptitle('Histogramas e Boxplots de Variáveis com Valores Ausentes', fontsize=16)
    
    for i, var in enumerate(variaveis):
        row = i // 2
        col_hist = (i % 2) * 2
        col_box = col_hist + 1
        
        sns.histplot(data[var].dropna(), bins=30, kde=True, ax=axs[row, col_hist])
        axs[row, col_hist].set_title(f'Histograma de {var}')
        
        sns.boxplot(x=data[var].dropna(), ax=axs[row, col_box])
        axs[row, col_box].set_title(f'Boxplot de {var}')
    
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()

# Plotar variáveis com valores ausentes
variaveis_ausentes = dados[dados.columns[dados.isna().any()]].columns.tolist()
plot_variaveis_ausentes(dados, variaveis_ausentes)

print(dados_temp.describe())
```

```python
from scipy.stats import mode

## Tratando com Imputação da Moda

# Função para imputar moda
def imputar_moda(data, columns):
    for column in columns:
        moda = mode(data[column].dropna())[0][0]
        data[column].fillna(moda, inplace=True)

# Variáveis para imputação da moda
variaveis_moda = ['Start_Month', 'Start_Day', 'End_Month', 'End_Day']
imputar_moda(dados, variaveis_moda)


## Tratando com Imputação da Mediana

# Função para imputar mediana
def imputar_mediana(data, columns):
    for column in columns:
        mediana = data[column].median()
        data[column].fillna(mediana, inplace=True)

# Variáveis para imputação da mediana
variaveis_mediana = ['Total_Deaths', 'Total_Affected']
imputar_mediana(dados, variaveis_mediana)


## Verificando Resultado

# Criar DataFrame temporário com as variáveis que possuem valores ausentes
dados_temp = dados[dados.columns[dados.isna().any()].tolist()]

# Exibir informações do DataFrame temporário
print(dados_temp.info())
print('\n\n================================================================================================\n\n')

# Exibir a quantidade de valores ausentes em cada variável
print('Total de Linhas Com Valores Ausentes Por Coluna\n\n')
print(dados_temp.isna().sum())
```

## Feature Engineering
> Vamos explorar algumas técnicas de Feature Engineering para criar novas variáveis que podem melhorar o desempenho do modelo.

- **Duração do Desastre (Disaster_Duration)**: A duração do desastre pode ser um indicativo da gravidade e do impacto do evento. Desastres mais longos podem resultar em maiores danos totais.

- **Lag Features (Lag_Total_Damages_1** e **Lag_Total_Damages_2)**: Os valores passados de danos podem influenciar os valores futuros, especialmente se houver um padrão de recorrência de desastres em certas regiões.

- **Variáveis Binárias (Is_Earthquake** e **Is_Flood**): Alguns tipos de desastres podem ter um impacto particularmente alto em danos, e criar variáveis binárias pode ajudar o modelo a capturar essas diferenças específicas.

- **Clusterização**: Identificando Regiões e Tipos de Desastres Semelhantes. A clusterização pode ajudar a identificar regiões que estão mais propensas a certos tipos de desastres, permitindo a criação de políticas e medidas preventivas específicas. Além disso, pode destacar padrões ocultos nos dados que não são facilmente visíveis através de análises tradicionais.

#### Duração do Desastre (Disaster_Duration)

```python
def ajustar_dia(mes, dia, ano):
    # Adjust for leap years
    if mes == 2:
        if (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0):  # Leap year check
            if dia > 29:
                dia = 29
        elif dia > 28:
            dia = 28
    elif mes in [4, 6, 9, 11] and dia > 30:
        dia = 30
    elif dia > 31:
        dia = 31
    return pd.to_datetime(f'{ano}-{mes:02d}-{dia:02d}', errors='coerce')

# Applying the function to correct invalid dates
dados['Start_Date'] = dados.apply(lambda row: ajustar_dia(int(row['Start_Month']), int(row['Start_Day']), int(row['Year'])), axis=1)
dados['End_Date'] = dados.apply(lambda row: ajustar_dia(int(row['End_Month']), int(row['End_Day']), int(row['Year'])), axis=1)

# Ensure End_Date is not earlier than Start_Date
dados['End_Date'] = dados.apply(lambda row: max(row['End_Date'], row['Start_Date']), axis=1)

# Calculate the duration of the disaster in days
dados['Disaster_Duration_Days'] = (dados['End_Date'] - dados['Start_Date']).dt.days

# Remove intermediate columns
dados.drop(columns=['Start_Date', 'End_Date'], inplace=True)

# Display statistical summary of Disaster_Duration_Days
print(dados['Disaster_Duration_Days'].describe())

# Visualização dos outliers antes do tratamento
plt.figure(figsize=(15, 10))
plt.subplot(1, 2, 1)
sns.boxplot(data=dados['Disaster_Duration_Days'])
plt.xticks(rotation=90)
plt.title('Boxplot para Detecção de Outliers (Antes do Tratamento)')

# Histograma antes do tratamento
plt.subplot(1, 2, 2)
sns.histplot(dados['Disaster_Duration_Days'], bins=50, kde=True)
plt.title('Histogram of Disaster Duration Days (Antes do Tratamento)')
plt.xlabel('Duration Days')
plt.ylabel('Frequency')
plt.show()

# Aplicar Winsorization
dados['Disaster_Duration_Days_Winsorized'] = winsorize(dados['Disaster_Duration_Days'], limits=[0.01, 0.01])

# Aplicar Transformação Logarítmica
dados['Disaster_Duration_Days_Log'] = np.log1p(dados['Disaster_Duration_Days_Winsorized'])

# Verificação de valores negativos ou NaN após a transformação
assert dados['Disaster_Duration_Days_Log'].isna().sum() == 0, "Existem valores NaN após a transformação logarítmica"
assert (dados['Disaster_Duration_Days_Log'] < 0).sum() == 0, "Existem valores negativos após a transformação logarítmica"

# Visualização dos outliers após Winsorization e Transformação Logarítmica
plt.figure(figsize=(15, 10))
plt.subplot(1, 2, 1)
sns.boxplot(data=dados['Disaster_Duration_Days_Log'])
plt.xticks(rotation=90)
plt.title('Boxplot para Detecção de Outliers (Após Tratamento)')

# Histograma após Winsorization e Transformação Logarítmica
plt.subplot(1, 2, 2)
sns.histplot(dados['Disaster_Duration_Days_Log'], bins=50, kde=True)
plt.title('Histogram of Disaster Duration Days (Após Tratamento)')
plt.xlabel('Log(Duration Days)')
plt.ylabel('Frequency')
plt.show()

# Estatísticas descritivas após tratamento
dados.drop(columns=['Disaster_Duration_Days', 'Disaster_Duration_Days_Winsorized'], inplace=True)
dados['Disaster_Duration_Days'] = dados['Disaster_Duration_Days_Log']
dados.drop(columns=['Disaster_Duration_Days_Log'], inplace=True)
print(dados['Disaster_Duration_Days'].describe())
```

### Clusterização

- A técnica de **clusterização cria uma nova variável** que representa os clusters identificados no conjunto de dados. Esta nova variável pode ser usada como uma feature adicional na criação de modelos preditivos, fornecendo informações sobre os padrões identificados durante a clusterização.

```python
## Preparação dos Dados

# Selecionar as variáveis relevantes para a clusterização
variables = ['Total_Deaths', 'Total_Affected', 'Total_Damages', 'Disaster_Duration_Days', 'Is_Storm', 'Is_Flood', 'Is_Earthquake']

# Criar um DataFrame apenas com as variáveis selecionadas
cluster_data = dados[variables]

# Normalizar os dados
scaler = StandardScaler()
cluster_data_scaled = scaler.fit_transform(cluster_data)


## Escolha do Algoritmo de Clusterização e Avaliação do Número Ótimo de Clusters

# Avaliar o método do cotovelo para encontrar o número ótimo de clusters
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(cluster_data_scaled)
    sse.append(kmeans.inertia_)

# Plotar o gráfico do método do cotovelo
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), sse, marker='o')
plt.title('Método do Cotovelo')
plt.xlabel('Número de Clusters')
plt.ylabel('SSE')
plt.show()
```

```python
## Execução da Clusterização e Criação da Nova Variável

# Escolher o número de clusters com base no gráfico do método do cotovelo (onde a cauda começa a abaixar)
optimal_clusters = 4

# Aplicar K-Means com o número ótimo de clusters
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
dados['Cluster'] = kmeans.fit_predict(cluster_data_scaled)

# Verificar as novas colunas criadas
print("\nNovas colunas criadas:")
print(dados.columns)

# Exibir uma amostra do DataFrame para verificar as novas colunas
display(dados.head(10))

# Visualizar os clusters resultantes
plt.figure(figsize=(15, 10))
sns.scatterplot(x='Total_Damages', y='Total_Affected', hue='Cluster', data=dados, palette='viridis')
plt.title('Clusterização de Regiões e Tipos de Desastres')
plt.xlabel('Total de Danos')
plt.ylabel('Total de Afetados')
plt.legend(title='Cluster')
plt.show()

# Mostrar a contagem de observações em cada cluster
print(dados['Cluster'].value_counts())

# Analisar as características médias de cada cluster
cluster_summary = dados.groupby('Cluster')[variables].mean()
display(cluster_summary)
```

```python
from sklearn.metrics import silhouette_score

# Calcular a pontuação de silhueta
silhouette_avg = silhouette_score(cluster_data_scaled, dados['Cluster'])
print(f'Pontuação de Silhueta: {silhouette_avg:.2f}')
```

### Criação, Treinamento, Previsão e Avaliação do Modelo

```python
## Criação e Treinamento

# Cria o modelo
modelo_RL = LinearRegression()

# Treina o modelo
modelo_RL.fit(X_treino, y_treino)
print(modelo_RL)

print('\n=======================================================================================\n')


## Previsões

# Previsões com dados de teste
pred_v1 = modelo_RL.predict(X_teste)

# Imprime as 10 primeiras previsões
print(pred_v1[:10])

print('\n==========================================================================\n')

# Plot das previsões vs valores reais
plt.figure(figsize=(10, 8))
plt.scatter(x=y_teste, y=pred_v1, color='skyblue', edgecolors='black')
plt.xlabel('Valor Real de Y')
plt.ylabel('Valor Previsto de Y')
plt.show()


## Avaliações

print('\n=======================================================================================\n')
print('\nAvaliação do Modelo\n\n')

# MAE - Erro Médio Absoluto
mae = mean_absolute_error(y_teste, pred_v1)
print('MAE - Erro Médio Absoluto:', mae)

# MSE - Erro Quadrático Médio
mse = mean_squared_error(y_teste, pred_v1)
print('MSE - Erro Quadrático Médio:', mse)

# RMSE - Raiz Quadrada do Erro Quadrático Médio
rmse = np.sqrt(mse)
print('RMSE - Raiz Quadrada do Erro Quadrático Médio:', rmse)

# Avaliação do RMSE em relação à média e ao desvio padrão
mean_value = dados['Total_Damages'].mean()
std_value = dados['Total_Damages'].std()
half_std_value = std_value / 2
rmse_as_percentage_of_mean = (rmse / mean_value) * 100
rmse_as_percentage_of_std = (rmse / std_value) * 100

if rmse < half_std_value:
    print(f'O valor é {rmse:.2f} e está MENOR que a metade do desvio padrão dos valores da variável alvo ({half_std_value:.2f}). Isso é BOM sinal.')
else:
    print(f'O valor é {rmse:.2f} e está MAIOR que a metade do desvio padrão dos valores da variável alvo ({half_std_value:.2f}). Isso NÃO é um bom sinal.')

# Coeficiente R2
r2 = r2_score(y_teste, pred_v1)
print('Coeficiente R2:', r2)

# Variância Explicada
evs = explained_variance_score(y_teste, pred_v1)
print('Variância Explicada:', evs)


print('\n==========================================================================\n')
print('ANÁLISE DE RESÍDUOS')


# Plot dos resíduos
residuos = y_teste - pred_v1

# Histograma dos resíduos
plt.figure(figsize=(10, 6))
sns.histplot(residuos, kde=True, color='purple')
plt.xlabel('Resíduos')
plt.title('Histograma dos Resíduos')
plt.show()
```

```python
# Cria o modelo Ridge com o melhor hiperparâmetro encontrado
modelo_v3_RR = Ridge(alpha=melhor_param_v3['alpha'])

# Treina o modelo
modelo_v3_RR.fit(X_treino_selecionados, y_treino_selecionados)

# Previsões com dados de teste
pred_v3 = modelo_v3_RR.predict(X_teste_selecionados)

# Imprime as 10 primeiras previsões
print(pred_v3[:10])

print('\n==========================================================================\n')

# Plot das previsões vs valores reais
plt.figure(figsize=(10, 8))
plt.scatter(x=y_teste_selecionados, y=pred_v3, color='skyblue', edgecolors='black')
plt.xlabel('Valor Real de Y')
plt.ylabel('Valor Previsto de Y')
plt.show()

## Avaliação do Modelo

# MAE - Erro Médio Absoluto
mae = mean_absolute_error(y_teste_selecionados, pred_v3)
print('MAE - Erro Médio Absoluto:', mae)

# MSE - Erro Quadrático Médio
mse = mean_squared_error(y_teste_selecionados, pred_v3)
print('MSE - Erro Quadrático Médio:', mse)

# RMSE - Raiz Quadrada do Erro Quadrático Médio
rmse = np.sqrt(mse)
print('RMSE - Raiz Quadrada do Erro Quadrático Médio:', rmse)

# Avaliação do RMSE em relação à média e ao desvio padrão
mean_value = dados['Total_Damages'].mean()
std_value = dados['Total_Damages'].std()
half_std_value = std_value / 2
rmse_as_percentage_of_mean = (rmse / mean_value) * 100
rmse_as_percentage_of_std = (rmse / std_value) * 100

if rmse < half_std_value:
    print(f'O valor é {rmse:.2f} e está MENOR que a metade do desvio padrão dos valores da variável alvo ({half_std_value:.2f}). Isso é BOM sinal.')
else:
    print(f'O valor é {rmse:.2f} e está MAIOR que a metade do desvio padrão dos valores da variável alvo ({half_std_value:.2f}). Isso NÃO é um bom sinal.')

# Coeficiente R2
r2 = r2_score(y_teste_selecionados, pred_v3)
print('Coeficiente R2:', r2)

# Variância Explicada
evs = explained_variance_score(y_teste_selecionados, pred_v3)
print('Variância Explicada:', evs)

print('\n==========================================================================\n')
print('ANÁLISE DE RESÍDUOS')

# Plot dos resíduos
residuos = y_teste_selecionados - pred_v3

# Histograma dos resíduos
plt.figure(figsize=(10, 6))
sns.histplot(residuos, kde=True, color='purple')
plt.xlabel('Resíduos')
plt.title('Histograma dos Resíduos')
plt.show()
```

### Função para Avaliar o Modelo

```python
def avaliar_modelo(nome_modelo, versao, tipo_modelo, modelo, X_treino, y_treino, X_teste, y_teste):
    # Treina o modelo
    modelo.fit(X_treino, y_treino)
    
    # Previsões com dados de teste
    pred = modelo.predict(X_teste)
    
    # Avaliações
    mae = mean_absolute_error(y_teste, pred)
    mse = mean_squared_error(y_teste, pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_teste, pred)
    evs = explained_variance_score(y_teste, pred)
    
    # Print das métricas
    print(f'\nAvaliação do {nome_modelo} - Versão {versao}\n')
    print('MAE - Erro Médio Absoluto:', mae)
    print('MSE - Erro Quadrático Médio:', mse)
    print('RMSE - Raiz Quadrada do Erro Quadrático Médio:', rmse)
    print('Coeficiente R2:', r2)
    print('Variância Explicada:', evs)
    
    # Plot das previsões vs valores reais
    plt.figure(figsize=(10, 8))
    plt.scatter(x=y_teste, y=pred, color='skyblue', edgecolors='black')
    plt.xlabel('Valor Real de Y')
    plt.ylabel('Valor Previsto de Y')
    plt.title(f'{nome_modelo} - Versão {versao}')
    plt.show()
    
    # Plot dos resíduos
    residuos = y_teste - pred
    plt.figure(figsize=(10, 6))
    sns.histplot(residuos, kde=True, color='purple')
    plt.xlabel('Resíduos')
    plt.title(f'Histograma dos Resíduos - {nome_modelo} - Versão {versao}')
    plt.show()
    
    return {
        'Nome do Modelo': nome_modelo,
        'Versao': versao,
        'Tipo de Dados': 'Variáveis Selecionadas',
        'Tipo de Modelo': tipo_modelo,
        'MAE': mae,
        'MSE': mse,
        'RMSE': rmse,
        'Coeficiente R2': r2,
        'Variância Explicada': evs
    }
```
