# Projeto: Previsão do Sucesso de Músicas na Indústria Musical

### Contexto da Vida Real

- A indústria musical é altamente competitiva e dinâmica. Com o avanço da tecnologia, tornou-se possível analisar grandes volumes de dados para entender tendências e classificar músicas em diferentes gêneros.
- Este projeto tem como **objetivo desenvolver um modelo preditivo que possa classificar músicas em diferentes gêneros com base em suas características acústicas e métricas associadas**.

### Objetivo do Projeto

- Desenvolver um modelo de Machine Learning capaz de classificar uma música em um gênero específico, utilizando um conjunto de dados real contendo informações sobre músicas.


### Perguntas de Negócio Adicionais

- 1) **Quais características acústicas têm maior influência na classificação de uma música em um gênero específico??**
> Entender quais atributos, como tempo, energia, dança, etc., mais impactam a classificação pode ajudar artistas e produtores a focarem em aspectos específicos durante a criação de novas músicas.

- 2) **Como as características das músicas variam entre diferentes gêneros?**
> Analisar a relação entre gênero musical e características acústicas pode fornecer insights sobre as peculiaridades de cada gênero musical.

- 3) **Existe uma correlação entre a duração de uma música e seu gênero?**
> Investigar se músicas mais curtas ou mais longas tendem a pertencer a certos gêneros pode ajudar na composição de músicas que atendam melhor a esses gêneros.

- 4) **Como as tendências musicais mudaram ao longo do tempo?**
> Analisar como as características das músicas em diferentes gêneros mudaram ao longo das décadas pode fornecer insights valiosos sobre a evolução dos gêneros musicais.

- 5) **Qual é o impacto da presença de artistas em músicas colaborativas na classificação de gêneros?**
> Explorar se músicas com múltiplos artistas tendem a pertencer a certos gêneros pode ajudar a formular estratégias de marketing e colaborações artísticas.

- 6) **Qual a relação entre a data de lançamento de uma música e seu gênero?**
> Analisar se certas épocas do ano tendem a resultar em músicas de determinados gêneros pode ajudar na programação de lançamentos.

- 7) **Como as avaliações dos usuários influenciam a classificação das músicas em gêneros?**
> Investigar se existe uma correlação entre avaliações positivas dos usuários (se disponível) e o gênero das músicas pode ajudar a entender a importância do feedback dos fãs.

- 8) **Quais são as características comuns entre as músicas que se destacam em cada gênero?**
> Identificar padrões entre as músicas que frequentemente se destacam em cada gênero pode ajudar na criação de músicas que têm maior probabilidade de sucesso dentro desses gêneros.

### Tratando Valores Outliers

<br>

> **Sugestão**:
É mais apropriado tratar cada variável **individualmente**, pois cada uma possui uma distribuição e comportamento diferente. Tratando-as individualmente, podemos aplicar métodos específicos para cada caso, o que resulta em uma limpeza mais precisa e adequada para o modelo.

<br>

### Tipos de Tratamento de Outliers

#### Isolation Forest

- **Vantagens**: Lida com alta dimensionalidade, não assume distribuição dos dados.
- **Desvantagens**: Sensível aos parâmetros, difícil de interpretar.

#### Método IQR (Interquartile Range)

- **Vantagens**: Simples de implementar e interpretar, eficaz para distribuições simétricas.
- **Desvantagens**: Não adequado para distribuições assimétricas ou dados de alta dimensionalidade.

#### Métodos Particulares

- **Winsorização**: Limita valores outliers a um certo percentil.
- **Remoção de Outliers**: Exclui completamente as observações outliers.
- **Transformações**: Aplica transformações (log, sqrt, box-cox) para reduzir o impacto dos outliers.
- **Substituição com Médias/Medianas**: Substitui outliers por valores médios ou medianos.

#### Outros Métodos Avançados

- **Clusterização (e.g., DBSCAN)**: Identifica outliers como pontos fora de clusters densos.
- **Análise de Componentes Principais (PCA)**: Identifica outliers em um espaço de componentes principais reduzido.
- **Modelos Baseados em Distribuições**: Modela a distribuição dos dados e identifica pontos com baixa probabilidade.

<br>

# Clusterização

- A técnica de **clusterização cria uma nova variável** que representa os clusters identificados no conjunto de dados. Esta nova variável pode ser usada como uma feature adicional na criação de modelos preditivos, fornecendo informações sobre os padrões identificados durante a clusterização.
- Verificar **variáveis relevantes** para a clusterização.
- Verificar no gráfico de cotovelo o valor de kmeans.
- **Não usar a variável alvo**.

<br>

#### Motivo das Variáveis Não Selecionadas para Clusterização
- **key**: Embora key represente a tonalidade da música, é uma variável categórica que pode não contribuir significativamente para a identificação de clusters baseados em características acústicas contínuas. Além disso, a tonalidade pode variar amplamente dentro dos gêneros, não sendo um fator de agrupamento primordial.
- **mode**: mode é uma variável binária que indica se a música está em uma escala maior ou menor. Similar ao key, a escala não fornece informações suficientes para distinguir grupos de músicas em um contexto de clusterização baseado em características acústicas contínuas.

<br>

#### Preparação dos Dados

```python
## Preparação dos Dados

# Selecionar as variáveis relevantes para a clusterização (excluindo a variável alvo 'genre')
variables = [
    'danceability', 'energy', 'loudness', 'speechiness', 
    'acousticness', 'instrumentalness', 'liveness', 
    'valence', 'tempo', 'duration_ms'
]

# Criar um DataFrame apenas com as variáveis selecionadas
cluster_data = dados[variables]

# Normalizar os dados
scaler = StandardScaler()
cluster_data_scaled = scaler.fit_transform(cluster_data)

## Escolha do Algoritmo de Clusterização e Avaliação do Número Ótimo de Clusters

# Avaliar o método do cotovelo para encontrar o número ótimo de clusters
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(cluster_data_scaled)
    sse.append(kmeans.inertia_)

# Plotar o gráfico do método do cotovelo
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), sse, marker='o')
plt.title('Método do Cotovelo')
plt.xlabel('Número de Clusters')
plt.ylabel('SSE')
plt.show()
```

#### Execução da Clusterização

```python
# Escolher o número de clusters com base no gráfico do método do cotovelo (onde a cauda começa a abaixar)
optimal_clusters = 3

# Aplicar K-Means com o número ótimo de clusters
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
dados['Cluster'] = kmeans.fit_predict(cluster_data_scaled)

# Verificar as novas colunas criadas
print("\nNovas colunas criadas:")
print(dados.columns)

# Exibir uma amostra do DataFrame para verificar as novas colunas
display(dados.head(10))

# Redução de Dimensionalidade com PCA para Visualização
# Aplicar PCA para reduzir os dados a duas dimensões
pca = PCA(n_components=2)
principal_components = pca.fit_transform(cluster_data_scaled)

# Visualizar os clusters resultantes usando os componentes principais
plt.figure(figsize=(15, 10))
sns.scatterplot(x=principal_components[:, 0], y=principal_components[:, 1], hue=dados['Cluster'], palette='viridis')
plt.title('Visualização dos Clusters com PCA')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.legend(title='Cluster')
plt.show()

# Mostrar a contagem de observações em cada cluster
print(dados['Cluster'].value_counts())

# Analisar as características médias de cada cluster
cluster_summary = dados.groupby('Cluster')[variables].mean()
display(cluster_summary)
```

#### Verificando Desempenho da Clusterização

```python
from sklearn.metrics import silhouette_score

# Calcular a pontuação de silhueta
silhouette_avg = silhouette_score(cluster_data_scaled, dados['Cluster'])
print(f'Pontuação de Silhueta: {silhouette_avg:.2f}')
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import joblib

# Carregar o modelo, scaler e informações do conjunto de dados
modelo_cat = joblib.load('modelo_catboost.pkl')
scaler = joblib.load('scaler.pkl')
dados_info = joblib.load('dados_info.pkl')

# Ordem das colunas esperada pelo modelo
ordem_colunas_modelo = dados_info['columns']

# Título da Aplicação
title = widgets.HTML("<h1>Previsão de Gênero Musical com CatBoost</h1>")
display(title)

# Campos de entrada para os valores das variáveis
danceability_input = widgets.FloatText(description='Danceability:', min=0.0, max=1.0)
energy_input = widgets.FloatText(description='Energy:', min=0.0, max=1.0)
key_input = widgets.IntText(description='Key:', min=0, max=11)
loudness_input = widgets.FloatText(description='Loudness:', min=-60.0, max=0.0)
speechiness_input = widgets.FloatText(description='Speechiness:', min=0.0, max=1.0)
acousticness_input = widgets.FloatText(description='Acousticness:', min=0.0, max=1.0)
instrumentalness_input = widgets.FloatText(description='Instrumentalness:', min=0.0, max=1.0)
liveness_input = widgets.FloatText(description='Liveness:', min=0.0, max=1.0)
valence_input = widgets.FloatText(description='Valence:', min=0.0, max=1.0)
tempo_input = widgets.FloatText(description='Tempo:', min=0.0, max=300.0)
duration_ms_input = widgets.IntText(description='Duration (ms):', min=0)
cluster_input = widgets.IntText(description='Cluster:', min=0, max=3)

inputs = [danceability_input, energy_input, key_input, loudness_input, speechiness_input, acousticness_input, instrumentalness_input, liveness_input, valence_input, tempo_input, duration_ms_input, cluster_input]

# Botão de Previsão
predict_button = widgets.Button(description="Prever Gênero Musical")
display(*inputs)
display(predict_button)

# Área de Exibição de Resultados
output = widgets.Output()
display(output)

# Função de Previsão
def predict_genre(b):
    with output:
        clear_output()
        # Coletar os valores de entrada
        values = [input_widget.value for input_widget in inputs]
        
        # Criação do DataFrame com a ordem correta das colunas
        data = pd.DataFrame([values], columns=ordem_colunas_modelo)
        
        # Padronização das variáveis contínuas
        variaveis_continuas = ordem_colunas_modelo[:11]  # As primeiras 11 colunas são contínuas
        data[variaveis_continuas] = scaler.transform(data[variaveis_continuas])
        
        # Previsão
        prediction = modelo_cat.predict(data)
        proba = modelo_cat.predict_proba(data)[0]
        
        # Exibição do resultado
        genre_labels = dados_info['label_encoders']['genre'].inverse_transform([int(prediction[0])])
        confidence = float(proba[prediction[0]]) * 100
        
        display(widgets.HTML(f"<h2>Este modelo calculou que há {confidence:.1f}% de chance do gênero musical ser {genre_labels[0]}.</h2>"))

predict_button.on_click(predict_genre)
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import joblib

# Carregar o modelo, scaler e informações do conjunto de dados
modelo_cat = joblib.load('modelo_catboost.pkl')
scaler = joblib.load('scaler.pkl')
dados_info = joblib.load('dados_info.pkl')

# Ordem das colunas esperada pelo modelo
ordem_colunas_modelo = dados_info['columns']

# Título da Aplicação
title = widgets.HTML("<h1>Previsão de Gênero Musical com CatBoost</h1>")
display(title)

# Dados das duas primeiras linhas do conjunto de dados original
dados_amostra_1 = {
    'danceability': 0.831,
    'energy': 0.814,
    'key': 2,
    'loudness': -7.364,
    'speechiness': 0.4200,
    'acousticness': 0.0598,
    'instrumentalness': 0.013400,
    'liveness': 0.0556,
    'valence': 0.3890,
    'tempo': 156.985,
    'duration_ms': 124539,
    'Cluster': 2
}

dados_amostra_2 = {
    'danceability': 0.850,
    'energy': 0.893,
    'key': 5,
    'loudness': -4.783,
    'speechiness': 0.0623,
    'acousticness': 0.0138,
    'instrumentalness': 0.000004,
    'liveness': 0.3720,
    'valence': 0.0391,
    'tempo': 218.050,
    'duration_ms': 98821,
    'Cluster': 0
}

# Função de Previsão
def predict_genre(dados):
    # Criação do DataFrame com a ordem correta das colunas
    data = pd.DataFrame([dados], columns=ordem_colunas_modelo)
    
    # Padronização das variáveis contínuas
    variaveis_continuas = ordem_colunas_modelo[:11]  # As primeiras 11 colunas são contínuas
    data[variaveis_continuas] = scaler.transform(data[variaveis_continuas])
    
    # Previsão
    prediction = modelo_cat.predict(data)
    proba = modelo_cat.predict_proba(data)[0]
    
    # Exibição do resultado
    genre_labels = dados_info['label_encoders']['genre'].inverse_transform([int(prediction[0])])
    confidence = float(proba[int(prediction[0])]) * 100
    
    return genre_labels[0], confidence

# Previsão para a primeira amostra
genre_1, confidence_1 = predict_genre(dados_amostra_1)
print(f"Amostra 1: Este modelo calculou que há {confidence_1:.1f}% de chance do gênero musical ser {genre_1}.")

# Previsão para a segunda amostra
genre_2, confidence_2 = predict_genre(dados_amostra_2)
print(f"Amostra 2: Este modelo calculou que há {confidence_2:.1f}% de chance do gênero musical ser {genre_2}.")

# Verificar se os dados estão sendo transformados corretamente
print("\nDados Amostra 1 Padronizados:")
data_amostra_1 = pd.DataFrame([dados_amostra_1], columns=ordem_colunas_modelo)
data_amostra_1[variaveis_continuas] = scaler.transform(data_amostra_1[variaveis_continuas])
print(data_amostra_1)

print("\nDados Amostra 2 Padronizados:")
data_amostra_2 = pd.DataFrame([dados_amostra_2], columns=ordem_colunas_modelo)
data_amostra_2[variaveis_continuas] = scaler.transform(data_amostra_2[variaveis_continuas])
print(data_amostra_2)
```
