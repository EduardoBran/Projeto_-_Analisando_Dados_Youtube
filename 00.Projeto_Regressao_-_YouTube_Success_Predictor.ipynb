#### user_created

- criar 4 novas colunas que representarão user_created na forma numérica

```python
# Função para limpar a string de data
def clean_date_string(date_str):
    date_str = date_str.replace('th', '').replace('st', '').replace('nd', '').replace('rd', '')
    return date_str

# Função para extrair dia, mês e ano de uma string de data no formato 'MMM DD, YYYY'
def extract_date_components(date_str):
    date_str = clean_date_string(date_str)
    date_obj = pd.to_datetime(date_str, format='%b %d, %Y', errors='coerce')
    if pd.isnull(date_obj):
        return pd.NA, pd.NA, pd.NA
    return date_obj.day, date_obj.month, date_obj.year

# Aplicar a função para criar novas colunas
dados_ml['dia_criacao_do_canal'], dados_ml['mes_criacao_do_canal'], dados_ml['ano_criacao_do_canal'] = zip(*dados_ml['data_criacao_do_canal'].apply(extract_date_components))


# Função para calcular a idade do canal
def calculate_channel_age(date_str):
    date_str = clean_date_string(date_str)
    date_obj = pd.to_datetime(date_str, format='%b %d, %Y', errors='coerce')
    if pd.isnull(date_obj):
        return pd.NA
    today = datetime.today()
    return (today.year - date_obj.year) - ((today.month, today.day) < (date_obj.month, date_obj.day))

# Aplicar a função para criar uma nova coluna de idade do canal
dados_ml['idade_do_canal'] = dados_ml['data_criacao_do_canal'].apply(calculate_channel_age)

# Remover a coluna user_created
dados_ml.drop(columns=['data_criacao_do_canal'], inplace=True)


# Verificar se ainda existem valores nulos nas novas colunas
print('\nVerificando valores nulos nas novas colunas:')
print(dados_ml[['dia_criacao_do_canal', 'mes_criacao_do_canal', 'ano_criacao_do_canal', 'idade_do_canal']].isnull().sum())

# Exibir as primeiras linhas para verificação
display(dados_ml.head())
```

```python
# Função para extrair a estimativa mínima dos ganhos mensais
def extract_min_estimated_earnings(earnings_str):
    # Remover caracteres não numéricos e espaços e ajustar K, M, e B
    earnings_str = earnings_str.replace('$', '').replace(',', '').replace(' - ', '-')
    
    # Pegar os dois valores
    earnings_range = earnings_str.split('-')
    
    def convert_to_float(value):
        if value == '':
            return 0.0
        if 'K' in value:
            return float(value.replace('K', '')) * 1000
        elif 'M' in value:
            return float(value.replace('M', '')) * 1000000
        elif 'B' in value:
            return float(value.replace('B', '')) * 1000000000
        else:
            return float(value)
    
    # Converter o valor mínimo para float
    min_earnings = convert_to_float(earnings_range[0].strip())
    
    return min_earnings

# Aplicar a função para criar a nova coluna
dados_ml['ganhos_mensais_estimados_minimo'] = dados_ml['ganhos_mensais_estimados'].apply(extract_min_estimated_earnings)


# Verificar as primeiras linhas para confirmação
display(dados_ml[['ganhos_mensais_estimados', 'ganhos_mensais_estimados_minimo']].head())


# Remover a coluna ganhos_mensais_estimados
dados_ml.drop(columns=['ganhos_mensais_estimados'], inplace=True)
```

### Criando Variáveis Temporais

```python
from datetime import datetime
import numpy as np

# Renomear colunas temporariamente
dados_ml.rename(columns={'ano_criacao_do_canal': 'year', 'mes_criacao_do_canal': 'month', 'dia_criacao_do_canal': 'day'}, inplace=True)

# Assumindo que a data atual é 2024-06-01 para cálculo
data_atual = datetime(2024, 6, 1)

# Criar a coluna data_ultima_publicacao
dados_ml['data_ultima_publicacao'] = pd.to_datetime(dados_ml[['year', 'month', 'day']], errors='coerce')


# Criar uma variável temporária substituindo valores 0 por 1 em idade_do_canal
dados_ml['idade_do_canal_temp'] = dados_ml['idade_do_canal'].replace(0, 1)

# Calcular as médias de vídeos publicados usando a variável temporária
dados_ml['media_videos_publicados_por_dia'] = dados_ml['total_de_videos_publicado'] / dados_ml['idade_do_canal_temp']
dados_ml['media_videos_publicados_por_mes'] = dados_ml['total_de_videos_publicado'] / (dados_ml['idade_do_canal_temp'] * 12)
dados_ml['media_videos_publicados_por_ano'] = dados_ml['total_de_videos_publicado'] / dados_ml['idade_do_canal_temp']

# Substituir NaNs resultantes da divisão por 0 ou valores inválidos
dados_ml['media_videos_publicados_por_dia'].fillna(0, inplace=True)
dados_ml['media_videos_publicados_por_mes'].fillna(0, inplace=True)
dados_ml['media_videos_publicados_por_ano'].fillna(0, inplace=True)

# Renomear as colunas de volta aos nomes originais
dados_ml.rename(columns={'year': 'ano_criacao_do_canal', 'month': 'mes_criacao_do_canal', 'day': 'dia_criacao_do_canal'}, inplace=True)

# Remover a coluna data_ultima_publicacao e a coluna temporária idade_do_canal_temp
dados_ml.drop(columns=['data_ultima_publicacao', 'idade_do_canal_temp'], inplace=True)

# Verificar valores nulos
print(dados_ml[['media_videos_publicados_por_dia', 'media_videos_publicados_por_mes', 'media_videos_publicados_por_ano']].isnull().sum())
```

# Clusterização

<br>

#### Preparação dos Dados

```python
# Seleção das variáveis relevantes
features = ['total_de_videos_publicado', 'total_de_inscritos', 'total_de_visualizacoes',
            'idade_do_canal', 'rank_social_blade', 'rank_inscritos', 'rank_visualizacoes',
            'ganhos_mensais_estimados_minimo', 'total_de_visualizacoes_ultimos_30_dias',
            'total_de_inscritos_ultimos_30_dias', 'media_crescimento_inscritos_mes']

# Padronização das variáveis
scaler = StandardScaler()
dados_ml_scaled = scaler.fit_transform(dados_ml[features])

# Adicionar as variáveis padronizadas ao DataFrame com um sufixo '_std'
dados_ml_std = pd.DataFrame(dados_ml_scaled, columns=[f"{col}_std" for col in features], index=dados_ml.index)
dados_ml = pd.concat([dados_ml, dados_ml_std], axis=1)

# Verificar a consistência dos índices
print(f"Shape of dados_ml: {dados_ml.shape}")
print(f"Shape of dados_ml_std: {dados_ml_std.shape}")

# Método do Cotovelo para determinar o número ideal de clusters
sse = []
k_range = range(1, 11)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(dados_ml_std)
    sse.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(k_range, sse, marker='o')
plt.title('Método do Cotovelo')
plt.xlabel('Número de Clusters')
plt.ylabel('SSE')
plt.show()
```

<br>

#### Execução da Clusterização

```python
# Número de clusters ideal (exemplo, assumindo que o cotovelo foi identificado em k=3)
k_optimal = 3
kmeans = KMeans(n_clusters=k_optimal, random_state=42)
dados_ml['Cluster'] = kmeans.fit_predict(dados_ml_std)

# Exibir o DataFrame resultante com a nova variável 'Cluster'
dados_ml.head()
```

```python
# Remover as variáveis padronizadas
dados_ml.drop(columns=[f"{col}_std" for col in features], inplace=True)
```

- Com Winsorização
- Todas as variáveis exceto 'tempo_desde_ultima_publicacao'.

```python
## Tratando Valores Outliers

# Função para aplicar Winsorização
def apply_winsorization(X, limits):
    return X.apply(lambda col: winsorize(col, limits=limits))

# Limites para Winsorização
winsor_limits = (0.05, 0.05)

# Obter as colunas de X_treino exceto a(s) indicada(s)
variables_to_winsorize = X_treino.columns.difference(['idade_do_canal'])

# Winsorização das variáveis numéricas contínuas nos dados de treino e teste
X_treino[variables_to_winsorize] = apply_winsorization(X_treino[variables_to_winsorize], winsor_limits)
X_teste[variables_to_winsorize] = apply_winsorization(X_teste[variables_to_winsorize], winsor_limits)

# Winsorização da variável alvo de treino e teste
y_treino = winsorize(y_treino, winsor_limits)
y_teste = winsorize(y_teste, winsor_limits)
```

### Função para Avaliar o Modelo

```python
# Função para avaliar modelo
def avaliar_modelo(nome_modelo, versao, tipo_modelo, modelo, X_treino, y_treino_scaled, X_teste, y_teste_scaled):
    # Treina o modelo
    modelo.fit(X_treino, y_treino_scaled.ravel())  # Transformar y_treino_scaled para 1D
    
    # Previsões com dados de teste
    pred_scaled = modelo.predict(X_teste)
    
    # Reverter a padronização das previsões e dos valores reais para a escala original
    pred = scaler_y.inverse_transform(pred_scaled.reshape(-1, 1))
    y_test = scaler_y.inverse_transform(y_teste_scaled.reshape(-1, 1))
    
    # Avaliações
    mae = mean_absolute_error(y_test, pred)
    mse = mean_squared_error(y_test, pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, pred)
    evs = explained_variance_score(y_test, pred)
    
    # Print das métricas
    print(f'\nAvaliação do {nome_modelo} - Versão {versao}\n')
    print('MAE - Erro Médio Absoluto:', mae)
    print('MSE - Erro Quadrático Médio:', mse)
    print('RMSE - Raiz Quadrada do Erro Quadrático Médio:', rmse)
    print('Coeficiente R2:', r2)
    print('Variância Explicada:', evs)
    
    # Plot das previsões vs valores reais
    plt.figure(figsize=(10, 8))
    plt.scatter(x=y_test, y=pred, color='skyblue', edgecolors='black')
    plt.xlabel('Valor Real de Y')
    plt.ylabel('Valor Previsto de Y')
    plt.title(f'{nome_modelo} - Versão {versao}')
    plt.show()
    
    # Plot dos resíduos
    residuos = y_test - pred
    plt.figure(figsize=(10, 6))
    sns.histplot(residuos, kde=True, color='purple')
    plt.xlabel('Resíduos')
    plt.title(f'Histograma dos Resíduos - {nome_modelo} - Versão {versao}')
    plt.show()
    
    return {
        'Nome do Modelo': nome_modelo,
        'Versao': versao,
        'Tipo de Dados': 'Reduzidos',
        'Tipo de Modelo': tipo_modelo,
        'MAE': mae,
        'MSE': mse,
        'RMSE': rmse,
        'Coeficiente R2': r2,
        'Variância Explicada': evs
    }
```

# Testando o Modelo Com Novos Dados

```python
import joblib
import pandas as pd
from scipy.stats.mstats import winsorize
from sklearn.preprocessing import StandardScaler
import numpy as np

# Função para aplicar Winsorização
def apply_winsorization(X, limits):
    return X.apply(lambda col: winsorize(col, limits=limits))

# Carregar os parâmetros salvos e o modelo para novos dados
scaler_X = joblib.load('pipeline_projeto_youtube/scaler_X.pkl')
scaler_y = joblib.load('pipeline_projeto_youtube/scaler_y.pkl')
modelo_GBR_v2 = joblib.load('pipeline_projeto_youtube/modelo_XGB_v2.pkl')
winsor_limits = joblib.load('pipeline_projeto_youtube/winsor_limits.pkl')
label_encoders = joblib.load('pipeline_projeto_youtube/label_encoders.pkl')
X_treino_columns = joblib.load('pipeline_projeto_youtube/X_treino_columns.pkl')

# Carregar os dados de mapeamento salvos
dados_ml = pd.read_csv('pipeline_projeto_youtube/dados_mapeamento.csv')

# Função para obter valores automáticos
def obter_valores_automaticos(pais):
    valores_pais = dados_ml[dados_ml['pais'] == pais].iloc[0]
    populacao_2022 = valores_pais['populacao_2022']
    total_usuarios_estimado_por_pais = valores_pais['total_usuarios_estimado_por_pais']
    return populacao_2022, total_usuarios_estimado_por_pais

# Função para calcular variáveis derivadas
def calcular_variaveis_derivadas(total_de_inscritos, total_de_visualizacoes, idade_do_canal,
                                 total_de_visualizacoes_ultimos_30_dias, total_de_inscritos_ultimos_30_dias,
                                 populacao_2022, total_usuarios_estimado_por_pais):
    
    proporcao_de_visualizacoes_recentes = total_de_visualizacoes_ultimos_30_dias / total_de_visualizacoes
    media_crescimento_inscritos_mes = total_de_inscritos_ultimos_30_dias / (idade_do_canal * 12)
    
    return proporcao_de_visualizacoes_recentes, media_crescimento_inscritos_mes

# Função para preparar os dados para o modelo
def preparar_dados_para_modelo(total_de_inscritos, total_de_visualizacoes, pais, idade_do_canal,
                               total_de_visualizacoes_ultimos_30_dias, total_de_inscritos_ultimos_30_dias):
    
    # Obter valores automáticos
    populacao_2022, total_usuarios_estimado_por_pais = obter_valores_automaticos(pais)
    
    # Calcular variáveis derivadas
    proporcao_de_visualizacoes_recentes, media_crescimento_inscritos_mes = calcular_variaveis_derivadas(
        total_de_inscritos, total_de_visualizacoes, idade_do_canal, total_de_visualizacoes_ultimos_30_dias,
        total_de_inscritos_ultimos_30_dias, populacao_2022, total_usuarios_estimado_por_pais
    )
    
    # Criar DataFrame com os novos dados
    novos_dados = pd.DataFrame([[
        total_de_inscritos, total_de_visualizacoes, pais, idade_do_canal,
        total_de_visualizacoes_ultimos_30_dias, total_de_inscritos_ultimos_30_dias,
        total_usuarios_estimado_por_pais, populacao_2022,
        proporcao_de_visualizacoes_recentes, media_crescimento_inscritos_mes
    ]], columns=X_treino_columns)
    
    # Aplicar Label Encoding para a variável categórica 'pais'
    novos_dados['pais'] = label_encoders['pais'].transform(novos_dados['pais'])
    
    #display(novos_dados)
    
    # Aplicar Winsorização
    variables_to_winsorize = novos_dados.columns.difference(['idade_do_canal'])
    novos_dados[variables_to_winsorize] = apply_winsorization(novos_dados[variables_to_winsorize], winsor_limits)
    
    # Separar colunas não escaladas
    non_scaled_columns = ['idade_do_canal']
    novos_dados_non_scaled = novos_dados[non_scaled_columns]
    novos_dados_to_scale = novos_dados.drop(columns=non_scaled_columns)
    
    # Aplicar padronização
    novos_dados_scaled = scaler_X.transform(novos_dados_to_scale)
    novos_dados_scaled = pd.DataFrame(novos_dados_scaled, columns=novos_dados_to_scale.columns, index=novos_dados.index)
    
    # Reconstruir DataFrame final
    novos_dados_final = pd.concat([novos_dados_scaled, novos_dados_non_scaled], axis=1)
    
    # Reordenar as colunas para garantir que estejam na mesma ordem que durante o treinamento
    novos_dados_final = novos_dados_final[X_treino_columns]
    
    return novos_dados_final

# Função para prever os ganhos mensais estimados
def prever_ganhos_mensais(novos_dados_final):
    # Prever usando o modelo carregado
    predicao = modelo_GBR_v2.predict(novos_dados_final)
    # Desfazer a padronização da variável alvo
    predicao_desnormalizada = scaler_y.inverse_transform(predicao.reshape(-1, 1))
    return predicao_desnormalizada[0][0]

# Exemplo de entrada do usuário
total_de_inscritos = 1060000.0
total_de_visualizacoes = 1.005169e+09
pais = 'Argentina'
idade_do_canal = 5  # anos
total_de_visualizacoes_ultimos_30_dias = 48085000.0
total_de_inscritos_ultimos_30_dias = 30000.0

# Preparar os dados para o modelo
novos_dados_final = preparar_dados_para_modelo(total_de_inscritos, total_de_visualizacoes, pais, idade_do_canal,
                                               total_de_visualizacoes_ultimos_30_dias, total_de_inscritos_ultimos_30_dias)

# Prever os ganhos mensais estimados
ganhos_mensais_estimados = prever_ganhos_mensais(novos_dados_final)

# Ajustar valores negativos para zero
if ganhos_mensais_estimados < 0:
    ganhos_mensais_estimados = 0 

print(f"Ganhos mensais estimados: U$$ {ganhos_mensais_estimados:,.2f}")
```

#### Aplicando Teste ANOVA para verificar se há diferenças significativas nos ganhos mensais entre os diferentes tipos de canal.

- Obrigatório Aplicar Label Encode antes

**Se o p-value é menor que 0.05**:
Contém Significância Estatística, ou seja **rejeitamos a hipótese nula**. Isso sugere que há uma diferença estatisticamente significativa entre os grupos (tipos de canal).

**Se o p-value é maior que 0.05**:
Sem Significância Estatística, ou seja **não rejeitamos a hipótese nula**. Isso sugere que não há uma diferença estatisticamente significativa entre os grupos (tipos de canal).

```python
from scipy.stats import f_oneway

# Grupos de ganhos por tipo de canal
grupos = dados_ml.groupby('tipo_do_canal')['ganhos_mensais_estimados_minimo'].apply(list)

# Execute ANOVA
anova_result = f_oneway(*grupos)
print(f'ANOVA F-statistic: {anova_result.statistic}, p-value: {anova_result.pvalue}')
```

<br>

#### Resposta

- Como o **p-value (8.09317571422026e-09) é muito menor que 0.05**, rejeitamos a hipótese nula. Isso sugere que há uma diferença estatisticamente significativa nos ganhos mensais entre os diferentes tipos de canal. Portanto o tipo de canal afeta significativamente os ganhos mensais no YouTube.

### 5. Como as tendências de ganhos dos canais variam ao longo do tempo e quais são as previsões futuras para canais específicos?

- A criação de uma **Análise Temporal** é crucial para responder de forma precisa e informada à pergunta sobre como as tendências de ganhos variam ao longo do tempo e quais são as previsões futuras para canais específicos, proporcionando uma visão clara do comportamento passado e futuro dos ganhos de um canal.

<br>

## Análise Simples

<br>

#### Preparando os Dados

```python
# Passo 1: Copiar dados_ml para um novo conjunto de dados
dados_simples = dados_ml.copy()

# Passo 2: Criar a coluna de data
dados_simples['data'] = pd.to_datetime(
    dados_simples[['ano_criacao_do_canal',
                   'mes_criacao_do_canal',
                   'dia_criacao_do_canal']].
    rename(columns={'ano_criacao_do_canal': 'year', 'mes_criacao_do_canal': 'month', 'dia_criacao_do_canal': 'day'}))


# Passo 3: Preparar os dados para Prophet
dados_simples.rename(columns={'data': 'ds', 'ganhos_mensais_estimados_minimo': 'y'}, inplace=True)

# Passo 4: Dividir os dados em treino e teste
train = dados_simples[dados_simples['ds'] < '2023-01-01']
test = dados_simples[dados_simples['ds'] >= '2023-01-01']
```

<br>

#### Treinando o Modelo e Visualizando Resultados

```python
## Criando e Treiando o Modelo

# Passo 5: Treinar o modelo Prophet
model = Prophet()
model.fit(train[['ds', 'y']])

# Passo 6: Criar dataframe para previsões futuras
future = model.make_future_dataframe(periods=len(test), freq='M')


## Previsões

# Passo 7: Fazer previsões
forecast = model.predict(future)

print('\n-----------------------------------------------------------------------------------------------\n')
print('Visualizar previsões\n')

# Passo 8: Visualizar previsões
fig = model.plot(forecast)
plt.title('Previsão de Ganhos Mensais Futuros')
plt.xlabel('Data')
plt.ylabel('Ganhos Mensais Estimados')
plt.show()

print('\n-----------------------------------------------------------------------------------------------\n')
print('Visualizar componentes das previsões (tendência, sazonalidade)\n')

# Passo 9: Visualizar componentes das previsões (tendência, sazonalidade)
fig2 = model.plot_components(forecast)
plt.show()
```

```python
## Avaliação do Modelo

# Passo 10: Avaliar o modelo comparando previsões com dados reais
plt.figure(figsize=(10, 6))
plt.plot(train['ds'], train['y'], label='Treino')
plt.plot(test['ds'], test['y'], label='Teste')
plt.plot(forecast['ds'], forecast['yhat'], label='Previsão')
plt.xlabel('Data')
plt.ylabel('Ganhos Mensais Estimados')
plt.title('Comparação entre Previsões e Dados Reais')
plt.legend()
plt.show()

# Passo 11: Calcular MAE e RMSE
mae = mean_absolute_error(test['y'], forecast['yhat'][-len(test):])
rmse = mean_squared_error(test['y'], forecast['yhat'][-len(test):], squared=False)

print('\n-----------------------------------------------------------------------------------------------\n')
print('Visualizar Métricas de Avaliação\n')

print(f'MAE: {mae}')
print(f'RMSE: {rmse}')
```

## Análise com Variáveis Adicionais

```python

# Passo 1: Copiar dados_ml para um novo conjunto de dados
dados_com_variaveis_adicionais = dados_ml.copy()

# Passo 2: Criar a coluna de data
dados_com_variaveis_adicionais['data'] = pd.to_datetime(dados_com_variaveis_adicionais[['ano_criacao_do_canal', 'mes_criacao_do_canal', 'dia_criacao_do_canal']].rename(columns={'ano_criacao_do_canal': 'year', 'mes_criacao_do_canal': 'month', 'dia_criacao_do_canal': 'day'}))

# Passo 3: Preparar os dados para Prophet
dados_com_variaveis_adicionais.rename(columns={'data': 'ds', 'ganhos_mensais_estimados_minimo': 'y'}, inplace=True)

# Incluir outras variáveis relevantes
dados_com_variaveis_adicionais['total_de_visualizacoes_ultimos_30_dias'] = dados_com_variaveis_adicionais['total_de_visualizacoes_ultimos_30_dias'].astype(float)
dados_com_variaveis_adicionais['total_de_inscritos_ultimos_30_dias'] = dados_com_variaveis_adicionais['total_de_inscritos_ultimos_30_dias'].astype(float)

# Passo 4: Dividir os dados em treino e teste
train = dados_com_variaveis_adicionais[dados_com_variaveis_adicionais['ds'] < '2023-01-01']
test = dados_com_variaveis_adicionais[dados_com_variaveis_adicionais['ds'] >= '2023-01-01']

# Passo 5: Treinar o modelo Prophet com variáveis adicionais
model = Prophet()
model.add_regressor('total_de_visualizacoes_ultimos_30_dias')
model.add_regressor('total_de_inscritos_ultimos_30_dias')
model.fit(train[['ds', 'y', 'total_de_visualizacoes_ultimos_30_dias', 'total_de_inscritos_ultimos_30_dias']])

# Passo 6: Criar dataframe para previsões futuras
future = model.make_future_dataframe(periods=len(test), freq='M')

# Adicionar valores futuros para os regressores (aqui usamos os mesmos valores históricos como exemplo)
future['total_de_visualizacoes_ultimos_30_dias'] = dados_com_variaveis_adicionais['total_de_visualizacoes_ultimos_30_dias']
future['total_de_inscritos_ultimos_30_dias'] = dados_com_variaveis_adicionais['total_de_inscritos_ultimos_30_dias']

# Passo 7: Fazer previsões
forecast = model.predict(future)

# Passo 8: Visualizar previsões
fig = model.plot(forecast)
plt.title('Previsão de Ganhos Mensais Futuros com Variáveis Adicionais')
plt.xlabel('Data')
plt.ylabel('Ganhos Mensais Estimados')
plt.show()

# Passo 9: Visualizar componentes das previsões (tendência, sazonalidade)
fig2 = model.plot_components(forecast)
plt.show()

# Passo 10: Avaliar o modelo comparando previsões com dados reais
plt.figure(figsize=(10, 6))
plt.plot(train['ds'], train['y'], label='Treino')
plt.plot(test['ds'], test['y'], label='Teste')
plt.plot(forecast['ds'], forecast['yhat'], label='Previsão')
plt.xlabel('Data')
plt.ylabel('Ganhos Mensais Estimados')
plt.title('Comparação entre Previsões e Dados Reais com Variáveis Adicionais')
plt.legend()
plt.show()

# Passo 11: Calcular MAE e RMSE
mae = mean_absolute_error(test['y'], forecast['yhat'][-len(test):])
rmse = mean_squared_error(test['y'], forecast['yhat'][-len(test):], squared=False)

print(f'MAE: {mae}')
print(f'RMSE: {rmse}')
```
